Ques 1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.

- **Relative cell references **are the default in Excel. They change automatically when a formula is copied to another cell, maintaining the relative position between cells. There is no dollar ($) sign in the relative reference for the cell.

                      Example : Formula in C1: =A1+B1


                        Copied to C2 â†’ becomes =A2+B2

 - **Absolute Cell Reference** When copying or using AutoFill, there are times when the cell reference must stay the same. A column and/or row reference is kept constant using dollar signs. So, to get an absolute reference from a relative, we can use the dollar sign ($) characters.

                               Example: $A$1
              Does not change when copied or moved.
                          Both the row and column are fixed using $.

- **Mixed references combine** relative and absolute references. You can lock either the row or the column, depending on your needs.


               $A1 (column fixed, row relative)


               A$1 (row fixed, column relative)


              Fixes either the row or the column, but not both.
